### Specify file path

```python
# Specify datapath
data_path  = "%s/%s"   % (os.getcwd(), "../data/")
indir      = "%s%s" % (data_path, "reconstruction/")
outdir     = "%s%s" % (data_path, "atomtrace/")
print("hostname:", os.uname()[1])
import arrayfire as af
af.set_device(0)
```

### Specify experiments to run

```python
#Experiments to be run
exps = [5004]
```

### Load Data

```python
for exp_i in exps:
    for file in os.listdir(indir):
        if fnmatch.fnmatch(file, "exp%s*" % (str(exp_i))):
            print(file)
            reconstruction = sio.loadmat(indir+file, squeeze_me=True)
            obj_final = reconstruction["obj_final"].real
            break
```

### Prepare kernel

```python
kernel_size = 11
coord = np.arange(-(kernel_size//2),kernel_size//2+1)
[zz, yy, xx] = np.meshgrid(coord, coord, coord)
s1=1
r2 = xx**2 + yy**2 + zz**2
kernel = -1*((r2 - 3*(s1**2))/np.sqrt(2*np.pi*(s1**2))) * np.exp(-r2/(2*(s1**2)))
plt.figure()
plt.imshow(kernel[kernel_size//2,])
plt.colorbar()
```

### Filter

```python
obj_final = np.array(af.to_array(obj_final))
atoms = np.array(af.convolve3(af.to_array(obj_final), af.to_array(kernel))).real
```

### Before and after filtering

```python
min_val = np.min(obj_final[:])
max_val = np.max(obj_final[:])
print("min:", min_val)
print("max:", max_val)
min_val1 = np.min(atoms[:])
max_val1 = np.max(atoms[:])
print("min:", min_val1)
print("max:", max_val1)
```

```python
plt.figure()
atoms[atoms < 300]=0
plt.hist(atoms.ravel(), bins = 100, range = [300, max_val1])
```

```python
compare3DStack(obj_final, atoms, clim=(min_val,max_val))
```
